```python
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
font_paths = mpl.font_manager.findSystemFonts()
font_objects = mpl.font_manager.createFontList(font_paths)
font_names = [f.name for f in font_objects]
```

```python
print(font_paths)
print(font_names)
```

```python
plt.figure()
plt.xlabel('てすと')
plt.title('テスト')
```
